### Practical (4A)
### Aim: Demonstrate the purpose of feature scaling and show that featuring scaling does not affect the distribution of the data. (Standardization)

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("Social_Network_Ads.csv",usecols=["Age","EstimatedSalary","Purchased"])
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(df.drop("Purchased",axis=1),df["Purchased"],
                                                   test_size=0.3,random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.fit_transform(x_test)
```

```python
 scaler.mean_
```

```python
x_train
```

```python
x_train_scaled
```

```python
x_train_scaled = pd.DataFrame(x_train_scaled,columns=x_train.columns)
x_test_scaled = pd.DataFrame(x_test_scaled,columns=x_test.columns)
```

```python
x_train_scaled
```

```python
import matplotlib.pyplot as plt
fig, (ax1,ax2) = plt.subplots(ncols=2,figsize=(12,5))

ax1.scatter(x_train["Age"],x_train["EstimatedSalary"])
ax1.set_title("Before Scaling")

ax2.scatter(x_train_scaled["Age"],x_train_scaled["EstimatedSalary"],color="green")
ax2.set_title("After scaling")
```

```python
import seaborn as sns
fig, (ax1,ax2) = plt.subplots(ncols=2,figsize=(12,5))

ax1.set_title("Before Scaling")
sns.kdeplot(x_train["Age"],ax=ax1)
sns.kdeplot(x_train["EstimatedSalary"],ax=ax1)

ax2.set_title("After Scaling")
sns.kdeplot(x_train_scaled["Age"],ax=ax2)
sns.kdeplot(x_test_scaled["EstimatedSalary"],ax=ax2)
```
