### Feature Scaling using correlation

```python
x=df.iloc[:,1:]
y=df.iloc[:,0]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
x_train.shape, x_test.shape
```

```python
cor=x_train.corr()
```

```python
cor
```

```python
plt.figure(figsize=(12,10))
sns.heatmap(cor,annot=True,cmap=plt.cm.CMRmap_r)
```

```python
def correlation(dataset, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr
```

```python
corr_f=correlation(x_train,0.7)
len(set(corr_f))
```

```python
corr_f
```

```python
x_train.drop(corr_f,axis=1)
x_test.drop(corr_f,axis=1)
```

### Chi-squared Test

```python
df2=pd.read_csv("sunny.csv")
df2
```

```python
df2.info()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
df2['Outlook'] = le.fit_transform(df2['Outlook'])
df2['temp'] = le.fit_transform(df2['temp'])
df2['humidity']=le.fit_transform(df2['humidity'])
df2['windy'] = le.fit_transform(df2['windy'])
df2['play'] = le.fit_transform(df2['play'])
```

```python
df2
```

```python
x = df2.iloc[:,:-1]
y = df2.iloc[:,-1]
```

```python
## Perform chi2 test
### chi2 returns 2 values
### Fscore and the pvalue
from sklearn.feature_selection import chi2
f_p_values=chi2(x,y)
```

```python
f_p_values
```

```python
p_values=pd.Series(f_p_values[0])
p_values.index=x.columns
p_values
```
